```python
from pathlib import Path
import pandas as pd
import requests

NOTEBOOKS_DIR = Path.cwd()
BASE_DIR = NOTEBOOKS_DIR.parent
DATA_DIR = BASE_DIR / "data"
SILVER_DIR = DATA_DIR / "silver"

caminho_selic_silver = SILVER_DIR / "selic_diaria.csv"

df_selic = pd.read_csv(caminho_selic_silver)

df_selic["data"] = pd.to_datetime(df_selic["data"])

df_selic["data"].min(), df_selic["data"].max()
```

```python
data_min_selic = df_selic["data"].min()
data_max_selic = df_selic["data"].max()

data_min_selic, data_max_selic
```

```python
DATA_INICIAL = data_min_selic.strftime("%m-%d-%Y")
DATA_FINAL = data_max_selic.strftime("%m-%d-%Y")

DATA_INICIAL, DATA_FINAL
```

```python
def buscar_dolar_periodo(data_inicial: str, data_final: str) -> pd.DataFrame:
    """
    Busca as cotações do dólar (PTAX) no período informado usando a API do Banco Central.

    Parâmetros:
    - data_inicial: string no formato 'MM-DD-YYYY', ex: '01-01-2020'
    - data_final: string no formato 'MM-DD'YYYY', ex: '12-31-2024'

    Retorna:
    - DataFrame com as colunas devolvidas pela API (cotacaoCompra, cotacaoVenda, dataHoraCotacao, data)
    """
    base_url = (
        "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/"
        "odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)"
    )

    params = {
        "@dataInicial": f"'{data_inicial}'",
        "@dataFinalCotacao": f"'{data_final}'",
        "$top": 10000,
        "$format": "json"
    }

    response = requests.get(base_url, params=params, timeout=10)
    response.raise_for_status()

    data_json = response.json()
    registros = data_json.get("value", [])

    df = pd.DataFrame(registros)

    return df
```

```python
caminho_selic = SILVER_DIR / "selic_diaria.csv"
caminho_dolar = SILVER_DIR / "dolar_diario.csv"

df_selic = pd.read_csv(caminho_selic)
df_dolar = pd.read_csv(caminho_dolar)
```

```python
df_dolar_bronze = buscar_dolar_periodo(DATA_INICIAL, DATA_FINAL)
df_dolar_bronze.head(), df_dolar_bronze.shape
```

```python
df_dolar_bronze["dataHoraCotacao"] = pd.to_datetime(df_dolar_bronze["dataHoraCotacao"])

df_dolar_bronze["data"] = df_dolar_bronze["dataHoraCotacao"].dt.normalize()

df_dolar_bronze[["dataHoraCotacao", "data"]].head()
```

```python
df_dolar_bronze["data"] = pd.to_datetime(df_dolar_bronze["data"])
df_dolar_bronze["ano"] = df_dolar_bronze["data"].dt.year

df_dolar_bronze["data"].min(), df_dolar_bronze["data"].max(), df_dolar_bronze["data"].value_counts().sort_index()
```

```python
df_dolar_silver = df_dolar_bronze.copy()
df_dolar_silver = df_dolar_silver[["data", "cotacaoCompra", "cotacaoVenda"]]
df_dolar_silver = df_dolar_silver[["data", "cotacaoCompra", "cotacaoVenda"]]

caminho_dolar_silver = SILVER_DIR / "dolar_diario.csv"
df_dolar_silver.to_csv(caminho_dolar_silver, index=False)

caminho_dolar_silver
```

```python
df_dolar = pd.read_csv(caminho_dolar_silver)
df_dolar["data"] = pd.to_datetime(df_dolar["data"])

df_dolar["ano"] = df_dolar["data"].dt.year
df_dolar["ano"].value_counts().sort_index()
```

```python
df_dolar["mes"] = df_dolar["data"].dt.month
df_dolar[["data", "ano", "mes"]].head()
```

```python
df_selic["data"] = pd.to_datetime(df_selic["data"])
df_dolar["data"] = pd.to_datetime(df_dolar["data"])
```

```python
col_data_selic = "data"
col_valor_selic = "valor"

col_data_dolar = "data"
col_valor_dolar = "cotacaoVenda"
```

```python
df_selic["ano"] = df_selic[col_data_selic].dt.year
df_selic["mes"] = df_selic[col_data_selic].dt.month

df_dolar["ano"] = df_dolar[col_data_dolar].dt.year
df_dolar["mes"] = df_dolar[col_data_dolar].dt.month

df_selic[["data", "ano", "mes"]].head(), df_dolar[["data", "ano", "mes"]].head()
```

```python
df_selic_mensal = (
    df_selic
    .groupby(["ano", "mes"], as_index=False)
    .agg(
        media_selic=(col_valor_selic, "mean")
    )
)

df_selic_mensal["dia"] = 1
df_selic_mensal["data_referencia"] = pd.to_datetime({
    "year": df_selic_mensal["ano"],
    "month": df_selic_mensal["mes"],
    "day": df_selic_mensal["dia"],
    
})

df_selic_mensal = df_selic_mensal.sort_values("data_referencia")
df_selic_mensal.head()
```

```python
df_selic["data"].min(), df_selic["data"].max()
```

```python
df_dolar_mensal = (
    df_dolar
    .groupby(["ano", "mes"], as_index=False)
    .agg(
        media_dolar=(col_valor_dolar, "mean")
    )
)

df_dolar_mensal["dia"] = 1
df_dolar_mensal["data_referencia"] = pd.to_datetime({
    "year": df_dolar_mensal["ano"],
    "month": df_dolar_mensal["mes"],
    "day": df_dolar_mensal["dia"],
})

df_dolar_mensal = df_dolar_mensal.sort_values("data_referencia")
df_dolar_mensal.head()
```

```python
df_dolar["data"].min(), df_dolar["data"].max()
```

```python
df_dolar[["ano", "mes"]].drop_duplicates().sort_values(["ano", "mes"])
```

```python
df_dolar["data"].min(), df_dolar["data"].max()
```

```python
df_dolar["ano"] = df_dolar["data"].dt.year
df_dolar["ano"].value_counts().sort_index()
```

```python
df_gold = pd.merge(
    df_selic_mensal[["data_referencia", "media_selic"]],
    df_dolar_mensal[["data_referencia", "media_dolar"]],
    on="data_referencia",
    how="inner"
)
df_gold.head()
```

```python
df_gold["diferenca_selic_dolar"] = df_gold["media_selic"] - df_gold["media_dolar"]

df_gold["relacao_selic_dolar"] = df_gold["media_selic"] / df_gold["media_dolar"]

df_gold.head()
```

```python
df_gold[["media_selic", "media_dolar"]].describe()
df_gold[["media_selic", "media_dolar"]].head()
df_gold[["media_selic", "media_dolar"]].tail()
```

```python
df_growth = df_gold.sort_values("data_referencia").copy()

base_selic = df_growth["media_selic"].iloc[0]
base_dolar = df_growth["media_dolar"].iloc[0]

df_growth["selic_idx"] = df_growth["media_selic"] / base_selic * 100
df_growth["dolar_idx"] = df_growth["media_dolar"] / base_dolar *100

df_growth[["data_referencia", "media_selic", "media_dolar", "selic_idx", "dolar_idx"]].head()
```

```python
GOLD_DIR = DATA_DIR / "gold"
GOLD_DIR.mkdir(parents=True, exist_ok=True)

caminho_gold = GOLD_DIR / "indicadores_selic_dolar_mensal.csv"
df_gold.to_csv(caminho_gold, index=False)

caminho_gold
```
